# **Deep Learning in Stock Market Based on Long Short Term Memory Technique**


**Introduction**:

We aim to use a Long Short Term Memory ensemble method
with two input sequences, a sequence of daily features and
a second sequence of annual features, in order to predict the
next day closing price and make a better decision in trading, then evaluate the method by comparing it to other techniques.

**Why LSTM?**

-LSTM is considered as an improvement of Reccurrent
Neural Network which comes as a solution to vanishing

and exploding gradient, Bao et al(2017); Werbos et al. (1988); Schmidhuber et al. (1997);

-LSTM is suitable for sequences Sutskever et al. (2014);

-LSTM can store and retrieve information using its gates
Bengio et al. (2009);

-LSTM doesnâ€™t flow in a single way (unlike Neural
Networks);

-LSTM technique can distinguish between recent and early
examples Nelson et al. (2017).

**Methodology:**

Our model combines two feature categories in order to
transmit the different aspects of an equity to the network.
The first aspect considers historical daily stock price that
reflects its trend. OHLC (Open, High, Low and Close stock
prices), volume and trading indicators MACD, RSI and Signal
show a share evolution based mainly on supply and demand.
The second aspect consists of financial metrics that are part
of the financial analysis. The objective is to take advantage
of both technical analysis and fundamental-financial analysis.
